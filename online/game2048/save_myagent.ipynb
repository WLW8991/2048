```python
import numpy as np
import csv, os
import sys
sys.path.append("..")

from game2048.expectimax import board_to_move
from game2048.agents_test import myAgent
from game2048.game import Game
from game2048.displays import Display
```

```python
GAME_SIZE = 4
SCORE_TO_WIN = 2048
repeat_times = 10

game = Game(GAME_SIZE, SCORE_TO_WIN)
board = game.board
agent = myAgent(game, Display())
direction = agent.step()
board = game.move(direction)

direct={}
i=0
k=0
filename = 'training.csv'

OUT_SHAPE=(4,4)
CAND=16
map_table={2**i: i for i in range (1,CAND)}
map_table[0]=0

if os.path.exists(filename):
	head = True
else:
	head = False
	os.mknod(filename)
    

def grid_ohe(arr):
    ret=np.zeros(shape=OUT_SHAPE+(CAND,),dtype=bool)
    for r in range(OUT_SHAPE[0]):
        for c in range(OUT_SHAPE[1]):
            ret[r,c,map_table[arr[r,c]]]=1
    return ret
```

```python
with open(filename, "a") as csvfile:
	writer = csv.writer(csvfile)
	if not head:
		writer.writerow(["11","12","13","14",\
			         "21","22","23","24",\
			         "31","32","33","34",\
			         "41","42","43","44",\
			         "direction"])

	while i < repeat_times:

		game = Game(GAME_SIZE, SCORE_TO_WIN)
		board = game.board
		print('Repeat times:', i)

		while(game.end == 0):
			agent =myAgent(game, Display())
			direction = agent.step()
			board = game.board
			board = grid_ohe(board)

			key = str(k+1)
			k = k + 1
			direct[key] = [np.int32(board), np.int32(direction)]
			data = np.int32(np.append(board, direction))
			writer.writerow(data)

			if k%100 == 0:
				print(data)

			game.move(direction)

		i = i + 1
```
